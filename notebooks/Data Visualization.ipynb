```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import torch 
import torch.nn.functional as F
from torchvision.transforms.functional import to_pil_image
import numpy as np
from torchvision.utils import draw_segmentation_masks
```

# Util Function

```python
def plot_images_side_by_side(image_paths, titles, subtitle, figsize=(12, 4)):
    """
    Plot multiple images side by side.

    Parameters:
        image_paths (list): List of image file paths.
        titles (list, optional): List of titles for each image. If None, no titles will be displayed.
        figsize (tuple, optional): Figure size. Default is (12, 4).
    """
    num_images = len(image_paths)

    plt.figure(figsize=figsize)
    plt.suptitle(subtitle)

    for i, path in enumerate(image_paths, start=1):
        plt.subplot(1, num_images, i)
        img = mpimg.imread(path)
        plt.imshow(img)
        if titles:
            plt.title(titles[i-1])
        plt.axis('off')
    
    plt.show()
```

```python
def plot_mask_and_image(image, mask):
    image = np.array(image, dtype = np.uint8).transpose(2,0,1)
    mask = np.array(mask).reshape(1,512, 512)

    image = torch.from_numpy(image).type(torch.uint8)
    mask = torch.from_numpy(mask).type(torch.bool)

    drawn = draw_segmentation_masks(image, mask, alpha = 0.7, colors = (0, 255, 0))

    drawn = to_pil_image(drawn).resize(size = (256, 256))
    
    return drawn
```

# Images and Mask Overlays Visualization

```python
norm_img1 = "../data/processed/train/images/100.png"
norm_img2 = "../data/processed/train/images/200.png"
norm_img3 = "../data/processed/train/images/300.png"

mask_img1 = "../data/processed/train/masks/100.png"
mask_img2 = "../data/processed/train/masks/200.png"
mask_img3 = "../data/processed/train/masks/300.png"
```

```python
image1 = Image.open(norm_img1)
mask1 = Image.open(mask_img1)

image2 = Image.open(norm_img2)
mask2 = Image.open(mask_img2)

image3 = Image.open(norm_img3)
mask3 = Image.open(mask_img3)
```

```python
comb1 = plot_mask_and_image(image1, mask1)
comb2 = plot_mask_and_image(image2, mask2)
comb3 = plot_mask_and_image(image3, mask3)
```

```python
def plot_multi_image(images, title):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.imshow(images[0])
    plt.title(title[0])
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(images[1])
    plt.title(title[1])
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(images[2])
    plt.title(title[2])
    plt.axis('off')

    plt.show()
    
plot_multi_image([comb1, comb2, comb3], title = ["Overlaid Image"] * 3)
```
